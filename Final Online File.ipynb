```python
import requests
import subprocess
import json

# Step 1: Call ChatGPT API
def call_chatgpt(prompt):
    api_key = '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'
    headers = {
        'Authorization': f'Bearer {api_key}',
        'Content-Type': 'application/json',
    }
    
    data = {
        'model': 'gpt-4',
        'messages': [{'role': 'user', 'content': prompt}],
        'max_tokens': 100,
    }
    
    response = requests.post('https://api.openai.com/v1/chat/completions', headers=headers, json=data)
    response_json = response.json()
    
    if response_json and 'choices' in response_json and len(response_json['choices']) > 0:
        return response_json['choices'][0]['message']['content']
    else:
        return "No response from ChatGPT."

# Step 2: Use curl to send data to Google Sheets
def send_to_google_sheets(prompt, response):
    url = 'https://script.google.com/macros/s/AKfycbyrjJ9Vc5l9WYR9hWBKbqLQj3Ima-7uDn1T8nvmTd1mc2SNAvw3yMvPllFzv3Hxkkcs/exec'
    
    # Data payload
    payload = json.dumps({
        'prompt': prompt,
        'response': response
    })
    
    # Construct the curl command
    curl_command = [
        'curl',
        '-X', 'POST', url,
        '-H', 'Content-Type: application/json',
        '-d', payload
    ]
    
    # Execute the curl command
    subprocess.run(curl_command, check=True)

# Step 3: Main Function
def main():
    prompt = input("Enter your prompt for ChatGPT: ")
    chatgpt_response = call_chatgpt(prompt)
    print(f"ChatGPT Response: {chatgpt_response}")
    
    # Send the prompt and response to Google Sheets
    send_to_google_sheets(prompt, chatgpt_response)
    print("Data sent to Google Sheets successfully.")

if __name__ == "__main__":
    main()
```
